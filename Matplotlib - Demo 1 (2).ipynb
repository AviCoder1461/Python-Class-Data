```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(10)
y=x**2
```

```python
print(x,y)
```

```python
fig=plt.figure()
axes1=fig.add_axes([0,0,1,1])
axes1.plot(x,y)
```

```python
fig=plt.figure()
ax1=fig.add_axes([.0,0.0,.8,.8])
ax1.plot(x,y)
ax2=fig.add_axes([.2,.5,.2,.2])
ax2.plot(y,x)
```

```python
fig,axes=plt.subplots(nrows=2,ncols=1)
for a_x in axes:
    a_x.plot(x,y)

#axes.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=2,ncols=1)
axes[0].plot(x,y)
axes[1].plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(7,3))
axes[0].plot(x,y)
axes[1].plot(x,y)
plt.tight_layout()
```
